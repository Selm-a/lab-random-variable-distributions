# Lab | Data cleaning and wrangling

```python
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
customer_df = customer_df.copy()
```

```python
#We will start with removing outliers. So far, we have discussed different methods to remove outliers. Use the one you feel more comfortable with, define a function for that. Use the function to remove the outliers and apply it to the dataframe.
#Create a copy of the dataframe for the data wrangling.
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    return (df[col] >= lower_limit) & (df[col] <= upper_limit)
contintuous_cols = continuous.columns
mask = customer_df[contintuous_cols].apply(lambda col: remove_outliers(customer_df, col.name))
customer_df_no_outliers = customer_df[mask.all(axis=1)]
```

```python
customer_df_no_outliers
```

```python
customer_df_no_outliers.select_dtypes(exclude=['object'])
```

```python
customer_df_no_outliers.info()
```

```python
#Normalize the continuous variables. You can use any one method you want.
#Encode the categorical variables
# One hot to state
# Ordinal to coverage
# Ordinal to employmentstatus
# Ordinal to location code
# One hot to marital status
# One hot to policy type
# One hot to policy
# One hot to renew offercustomer_df
# One hot to sales channel
# One hot vehicle class
# Ordinal vehicle size
```

```python
numeric_features = [1,8,11, 12, 13, 14, 15] # Choose which column index we are going to scale
numeric_transformer = StandardScaler()

categorical_features = [0, 10, 16, 17, 18 , 19, 21, 22]
categorical_transformer = OneHotEncoder()


# Apply ColumnTransformer to create a pipeline that will apply the above preprocessing
feature_encoder = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features),
        ('num', numeric_transformer, numeric_features)
        ]
    )
```

```python
customer_df_no_outliers["coverage"] = customer_df_no_outliers["coverage"].map({"Basic" : 0, "Extended" : 1, "Premium" : 2})
customer_df_no_outliers["employment_status"] = customer_df_no_outliers["employment_status"].map({"Employed" : 0, "Unemployed" : 1, "Disabled" : 2, "Medical Leave" : 3, "Retired" : 4})
customer_df_no_outliers["location_code"] = customer_df_no_outliers["location_code"].map({"Suburban" : 0, "Rural" : 1, "Urban" : 2})
customer_df_no_outliers["vehicle_size"] = customer_df_no_outliers["vehicle_size"].map({"Medsize" : 0, "Small" : 1, "Large" : 2})

```

```python
#The time variable can be useful. Try to transform its data into a useful one. Hint: Day week and month as integers might be useful.
```

```python
customer_df_no_outliers["effective_to_date_day"] = pd.to_datetime(customer_df_no_outliers["effective_to_date"]).dt.day
customer_df_no_outliers["effective_to_date_month"] = pd.to_datetime(customer_df_no_outliers["effective_to_date"]).dt.month
```

```python
X = customer_df_no_outliers.iloc[:, :-1] # explanatory variables

y = customer_df_no_outliers.iloc[:,-1] #  target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state =0)
```

```python
X_train = feature_encoder.fit_transform(X_train)

X_test = feature_encoder.transform(X_test) 
```

```python
#Try a simple linear regression with all the data to see whether we are getting good results.
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
```

```python
def model_inplace(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train) 
    print("r2 score on train set", model.score(X_train, y_train))
    print("r2 score on test set", model.score(X_test, y_test))
 
```

```python
LIN = LinearRegression()
KNR = KNeighborsRegressor(n_neighbors=2)
MLP = MLPRegressor(random_state=1, max_iter=500)
```

```python
model_inplace(LIN, X_train, X_test, y_train, y_test)
```

```python
model_inplace(KNR, X_train, X_test, y_train, y_test)
```

```python
model_inplace(MLP, X_train, X_test, y_train, y_test)
```
